```python
import pandas as pd

df = pd.read_csv('../data/curated/learner_activity_curated.csv')
df.head()
```

```python
funnel_counts = (
    df['funnel_stage']
    .value_counts()
    .reset_index()
    .rename(columns={'index': 'funnel_stage', 'funnel_stage': 'learners'})
)

funnel_counts
```

```python
funnel_counts.columns
```

```python
funnel_counts = funnel_counts.rename(
    columns={
        'learners': 'funnel_stage',
        'count': 'learners'
    }
)

funnel_counts
```

```python
order = ['Enrolled', 'Started', '25%', '50%', '75%', 'Completed']

funnel_counts['funnel_stage'] = pd.Categorical(
    funnel_counts['funnel_stage'],
    categories=order,
    ordered=True
)

funnel_counts = funnel_counts.sort_values('funnel_stage')
funnel_counts
```
