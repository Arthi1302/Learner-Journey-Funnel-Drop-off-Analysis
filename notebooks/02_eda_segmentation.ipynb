```python
import pandas as pd

df = pd.read_csv('../data/cleaned/learner_activity_cleaned.csv')

df.head()
```

```python
def funnel_stage(progress):
    if progress >= 100:
        return "Completed"
    elif progress >= 75:
        return "75%"
    elif progress >= 50:
        return "50%"
    elif progress >= 25:
        return "25%"
    elif progress > 0:
        return "Started"
    else:
        return "Enrolled"

df['funnel_stage'] = df['video_progress_pct'].apply(funnel_stage)

df['funnel_stage'].value_counts()
```

```python
df['learner_type'] = df['time_spent_minutes'].apply(
    lambda x: 'Power Learner' if x >= 90 else 'Casual Learner'
)

df['learner_type'].value_counts()
```

```python
df['learning_pace' ] = df['time_spent_minutes'].apply(
    lambda x: 'Binge Learner' if x >= 120 else 'Slow Learner'
)

df['learning_pace'].value_counts()

```

```python
df.to_csv('../data/curated/learner_activity_curated.csv', index=False)
```
